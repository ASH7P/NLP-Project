```python
import pandas as pd

df = pd.read_parquet("hf://datasets/Omar-youssef/Egyptian-text-summarization/data/train-00000-of-00001.parquet")
```

```python
if "id" not in df.columns:
    df["id"] = range(len(df))

df["text"] = df["text"].fillna("").astype(str)
df["summarized_text"] = df["summarized_text"].fillna("").astype(str)
df["source_topics"] = df["source_topics"].fillna("").astype(str)

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print(df.head(3))
```

```python
import re

def clean_egyptian_arabic(text):
    if not isinstance(text, str):
        return ""

    text = text.strip()

    # Remove diacritics + tatweel
    text = re.sub(r'[\u064B-\u0652\u0640]', '', text)

    # Normalize Alef variants
    text = re.sub(r'[إأآ]', 'ا', text)

    # Normalize Ya
    text = re.sub(r'ى', 'ي', text)

    # Remove URLs (very rare but safe)
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)

    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df_egyptian = df
```

```python
df_egyptian['text'] = df_egyptian['text'].apply(clean_egyptian_arabic)
df_egyptian['summarized_text'] = df_egyptian['summarized_text'].apply(clean_egyptian_arabic)

df_egyptian['summarized_text'] = df_egyptian['summarized_text'].apply(lambda x: f'<start> {x} <end>')
```

```python
df_egyptian.to_csv('cleaned_egyptian_final.csv', index=False)
df_egyptian
```
